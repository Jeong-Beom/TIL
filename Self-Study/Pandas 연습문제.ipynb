```python
# 타이타닉호
titanic = sns.load_dataset('titanic')
titanic['category1'] = titanic.apply(lambda x: x.age if x.age >= 20 else 'child', axis = 1)
titanic_age_mean = titanic.age.describe()['mean']
titanic.sort_values(by = 'age', inplace = True)
titanic.age.fillna(titanic_age_mean, inplace = True)
titanic['category1'] = titanic.apply(lambda x: x.age if x.age >= 20 else 'child', axis = 1)
titanic.pivot_table('survived','sex')
titanic.pivot_table('survived','embark_town')
titanic.pivot_table('survived','pclass')
bins = [0, 20, 30, 50, 70, 100]
labels = ['미성년자', '청년', '중년', '장년', '노년']
cats = pd.cut(titanic.age, bins, labels = labels)
titanic['category2'] = cats
titanic.pivot_table('survived','category2')
titanic.sort_values(by = 'age', ascending = False, inplace = True)
titanic['age_group'] = pd.qcut(titanic.age, 3, labels=['1', '2', '3'])
titanic.pivot_table('survived','age_group')
```

```python
# mpg
mpg = sns.load_dataset('mpg')
mpg['hp_per_cc'] = mpg.horsepower / mpg.displacement
mpg.pivot_table('mpg', 'cylinders')
mpg.pivot_table('hp_per_cc', 'origin')
mpg['manufacturer'] = mpg.name.apply(lambda x: x.split()[0])
mpg['model'] = mpg.name.apply(lambda x: ' '.join(x.split()[1:])) 
mpg[['manufacturer', 'model']].sort_values(by = 'manufacturer')
```
