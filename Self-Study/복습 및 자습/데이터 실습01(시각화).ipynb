```python
import re
seoul_park.면적
re.sub('[가-힣a-zA-Z()m²㎡]', '', seoul_park.면적[1])
for i in range(len(seoul_park.면적)):
    try: 
        seoul_park.면적[i] = re.sub('[가-힣a-zA-Z()m²㎡㎥/× ]', '', seoul_park.면적[i])
    except:
        print(i)
        seoul_park.면적[i] = 0
seoul_park.면적.head(20)
```

```python
for i in range(len(seoul_park.면적)):
    try:
        slice_num = seoul_park.면적[i].find(':')
        if slice_num != -1:
            seoul_park.면적[i] = seoul_park.면적[i][:slice_num] 
    except:
        pass
'''
x = seoul_park.면적[0].replace('㎡', '/')
num = x.find('/')
seoul_park.면적[0] = seoul_park.면적[0][:num] 
seoul_park.면적[0]
'''
for i in range(len(seoul_park.면적)):
    try: 
        seoul_park.면적[i] = int(np.round(float(seoul_park.면적[i])))
    except:
        seoul_park.면적[i] = 0
        pass
seoul_park.면적
```
